```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = "../data:clean:/palworld_2024-12-09_to_2025-01-13.csv"
output_path = "../outputs:/palworld_players_trend.png"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

df = pd.read_csv(file_path)

df["DateTime"] = pd.to_datetime(df["DateTime"])
df = df.sort_values("DateTime")
df["MA_7"] = df["Players"].rolling(window=7, min_periods=1).mean()
df["EMA_7"] = df["Players"].ewm(span=7, adjust=False).mean()

# Build a simple linear trend line across days.
df["DayIndex"] = (df["DateTime"] - df["DateTime"].min()).dt.days
trend_coef = np.polyfit(df["DayIndex"], df["Players"], 1)
df["Trend"] = np.polyval(trend_coef, df["DayIndex"])

plt.figure(figsize=(10, 5))
plt.plot(df["DateTime"], df["Players"], marker="o", linewidth=2, label="Daily Players")
plt.plot(df["DateTime"], df["MA_7"], color="crimson", linewidth=2.5, label="7-Day Moving Average")
plt.plot(df["DateTime"], df["EMA_7"], color="purple", linewidth=2.5, label="7-Day EMA (Smoothed)")
plt.plot(df["DateTime"], df["Trend"], color="darkgreen", linestyle="--", linewidth=2, label="Linear Trend")
plt.axvline(x=pd.to_datetime("2024-12-23"), color="red", linestyle=":", linewidth=2, label="Update Date (2024-12-23)", alpha=0.7)
plt.title("Palworld Players Over Time (Clean Data)")
plt.xlabel("Date")
plt.ylabel("Players")
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig(output_path, dpi=300, bbox_inches="tight")
print(f"Saved plot to: {output_path}")
plt.show()
```
