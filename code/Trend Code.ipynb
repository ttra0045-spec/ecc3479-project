```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = "../data:clean/palworld_2024-12-09_to_2025-01-13.csv"
output_path = "../outputs/palworld_players_trend.png"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

df = pd.read_csv(file_path)

df["DateTime"] = pd.to_datetime(df["DateTime"])
df = df.sort_values("DateTime")
df["MA_7"] = df["Players"].rolling(window=7, min_periods=1).mean()
df["EMA_7"] = df["Players"].ewm(span=7, adjust=False).mean()

# Build a simple linear trend line across days.
df["DayIndex"] = (df["DateTime"] - df["DateTime"].min()).dt.days
trend_coef = np.polyfit(df["DayIndex"], df["Players"], 1)
df["Trend"] = np.polyval(trend_coef, df["DayIndex"])

plt.figure(figsize=(10, 5))
plt.plot(df["DateTime"], df["Players"], marker="o", linewidth=2, label="Daily Players")
plt.plot(df["DateTime"], df["MA_7"], color="crimson", linewidth=2.5, label="7-Day Moving Average")
plt.plot(df["DateTime"], df["EMA_7"], color="purple", linewidth=2.5, label="7-Day EMA (Smoothed)")
plt.plot(df["DateTime"], df["Trend"], color="darkgreen", linestyle="--", linewidth=2, label="Linear Trend")
plt.axvline(x=pd.to_datetime("2024-12-23"), color="red", linestyle=":", linewidth=2, label="Update Date (2024-12-23)", alpha=0.7)
plt.title("Palworld Players Over Time (Clean Data)")
plt.xlabel("Date")
plt.ylabel("Players")
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig(output_path, dpi=300, bbox_inches="tight")
print(f"Saved plot to: {output_path}")
plt.show()
```

```python
import pandas as pd
import numpy as np

def analyze_player_trend(file_path, update_date_str):
    # Load the dataset
    df = pd.read_csv(file_path)
    
    # Convert 'DateTime' to datetime objects
    df['DateTime'] = pd.to_datetime(df['DateTime'])
    df = df.sort_values('DateTime')
    
    # Define the update date
    update_date = pd.to_datetime(update_date_str)
    
    # Define the periods before and after the update
    before_update_start = update_date - pd.Timedelta(days=7)
    before_update_end = update_date - pd.Timedelta(days=1)
    after_update_start = update_date
    after_update_end = update_date + pd.Timedelta(days=6)
    
    # Filter the DataFrame for the periods before and after the update
    before_df = df[(df['DateTime'] >= before_update_start) & (df['DateTime'] <= before_update_end)]
    after_df = df[(df['DateTime'] >= after_update_start) & (df['DateTime'] <= after_update_end)]
    
    # Calculate the average number of players
    avg_players_before = before_df['Players'].mean()
    avg_players_after = after_df['Players'].mean()
    
    # Calculate the percentage change
    if avg_players_before > 0:
        percentage_change = ((avg_players_after - avg_players_before) / avg_players_before) * 100
    else:
        percentage_change = float('inf') # Handle division by zero
        
    # Calculate the linear trend for the entire period
    df['DayIndex'] = (df['DateTime'] - df['DateTime'].min()).dt.days
    trend_coef = np.polyfit(df['DayIndex'], df['Players'], 1)
    slope = trend_coef[0]
    
    return avg_players_before, avg_players_after, percentage_change, slope

# List of games to analyze
games = [
    {"name": "Palworld", "file": "../data:clean/palworld_2024-12-09_to_2025-01-13.csv", "update": "2024-12-23"},
    {"name": "Warframe", "file": "../data:clean/warframe_2025-11-26_to_2025-12-31.csv", "update": "2025-12-10"},
    {"name": "Counter-Strike 2", "file": "../data:clean/counter_strike_2_2024-04-23_to_2024-05-28.csv", "update": "2024-05-07"},
    {"name": "Sea of Thieves", "file": "../data:clean/sea_of_thieves_2024-10-03_to_2024-11-07.csv", "update": "2024-10-17"},
    {"name": "PUBG: BATTLEGROUNDS", "file": "../data:clean/pubg__battlegrounds_2025-10-22_to_2025-11-26.csv", "update": "2025-11-05"},
    {"name": "Dead by Daylight", "file": "../data:clean/dead_by_daylight_2023-11-14_to_2023-12-19.csv", "update": "2023-11-28"},
    {"name": "Don't Starve Together", "file": "../data:clean/don_t_starve_together_2023-04-13_to_2023-05-18.csv", "update": "2023-04-27"},
    {"name": "Deep Rock Galactic", "file": "../data:clean/deep_rock_galactic_2024-05-30_to_2024-07-04.csv", "update": "2024-06-13"},
    {"name": "Apex Legends", "file": "../data:clean/apex_legends_2025-01-28_to_2025-03-04.csv", "update": "2025-02-11"},
    {"name": "Destiny 2", "file": "../data:clean/destiny_2_2024-09-24_to_2024-10-29.csv", "update": "2024-10-08"},
    {"name": "No Man's Sky", "file": "../data:clean/no_man_s_sky_2025-01-15_to_2025-02-19.csv", "update": "2025-01-29"},
    {"name": "HELLDIVERS 2", "file": "../data:clean/helldivers_2_2025-08-19_to_2025-09-23.csv", "update": "2025-09-02"},
    {"name": "Path of Exile", "file": "../data:clean/path_of_exile_2025-06-06_to_2025-07-04.csv", "update": "2025-06-13"},
    {"name": "Rainbow Six Siege", "file": "../data:clean/rainbow_six_siege_2024-08-19_to_2024-09-16.csv", "update": "2024-08-26"},
    {"name": "The Finals", "file": "../data:clean/the_finals_2024-12-05_to_2025-01-02.csv", "update": "2024-12-12"},
    {"name": "V Rising", "file": "../data:clean/v_rising_2025-04-21_to_2025-05-19.csv", "update": "2025-04-28"}
]

# Run the analysis for each game and store the results
results = {}
for game in games:
    try:
        avg_before, avg_after, change, slope = analyze_player_trend(game["file"], game["update"])
        results[game["name"]] = {
            "avg_before": avg_before,
            "avg_after": avg_after,
            "percentage_change": change,
            "slope": slope
        }
    except FileNotFoundError:
        print(f"File not found for {game['name']}: {game['file']}")
    except Exception as e:
        print(f"An error occurred while analyzing {game['name']}: {e}")

# Print the results
for name, data in results.items():
    print(f"--- {name} ---")
    print(f"Average players before update: {data['avg_before']:.2f}")
    print(f"Average players after update: {data['avg_after']:.2f}")
    print(f"Percentage change: {data['percentage_change']:.2f}%")
    print(f"Linear trend slope: {data['slope']:.2f}\n")
```
