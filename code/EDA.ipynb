```python
import pandas as pd
import os

# Path to the directory containing the raw CSV files
data_path = '../data:clean:/'

# List all files in the directory
files = os.listdir(data_path)

# Filter for CSV files
csv_files = [file for file in files if file.endswith('.csv')]

# Apply the describe function to each CSV file
for csv_file in csv_files:
    file_path = os.path.join(data_path, csv_file)
    df = pd.read_csv(file_path)
    print(f"Description for {csv_file}:")
    print(df.describe())
    print('\\n' + '-'*50 + '\\n')
```

```python
from tabulate import tabulate
import os

# Create a dictionary to hold the descriptions
descriptions = {}

# Apply the describe function to each CSV file and store it
for csv_file in csv_files:
    file_path = os.path.join(data_path, csv_file)
    df = pd.read_csv(file_path)
    descriptions[csv_file] = df.describe()

# Now, let's format this for the markdown file
markdown_output = "## Clean Data Summary\n\n"
for game, desc_df in descriptions.items():
    markdown_output += f"### {game}\n\n"
    markdown_output += desc_df.to_markdown()
    markdown_output += "\n\n"

# Define the output path using a relative path from the Code directory
output_path = "../outputs:/clean_data_summary.md"

# Write the output to a file
with open(output_path, "w") as f:
    f.write(markdown_output)

print(f"Clean data summary has been written to {os.path.abspath(output_path)}")
```
